The goal is to predict the quality of the student. We will build a predictor based on the final grade (G3).
Becasue we are trying to find quality students. In this model we define a quality student as one who achieves a final grade of 15 or higher. 

```python
include = ['health','Medu','Fedu','studytime','traveltime','absences','age','G3']
train.drop(columns=train.columns.difference(include), inplace=True)  # only using 3 features
train.info
```

```python
test['qual_student'] = np.where(test['G3']>=15, 1, 0)
test.describe()
```

```python
train['qual_student'] = np.where(train['G3']>=15, 1, 0)
train.describe()
```

Drop the G3 score

```python
include = ['health','Medu','Fedu',
'studytime','traveltime','absences','age','qual_student']
test.drop(columns=test.columns.difference(include), inplace=True)
test.describe()
```

```python
include = ['health','Medu','Fedu',
'studytime','traveltime','absences','age','qual_student']
train.drop(columns=train.columns.difference(include), inplace=True)
train.describe()
```

Import scikit-learn and build a random forest classifer

```python
from sklearn.ensemble import RandomForestClassifier as rf
import sklearn
import os
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

dependent_variable = 'qual_student'

X_test = test[test.columns.difference([dependent_variable])]
Y_test = test[dependent_variable]

X_train = train[train.columns.difference([dependent_variable])]
Y_train = train[dependent_variable]

# Random Forest
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
random_forest_preds = random_forest.predict(X_test)
print('The accuracy of the Random Forests model is :\t',metrics.accuracy_score(random_forest_preds,Y_test))


#clf = rf(n_estimators = 1000)
#clf.fit(x_test, y_test)
```

Let's export this model so we can use it in a microservice (flask api)

```python
import lime
import lime.lime_tabular
```

```python
predict_fn_rf = lambda x: random_forest.predict_proba(x).astype(float)
X = X_train.values

explainer = lime.lime_tabular.LimeTabularExplainer(X,feature_names = X_train.columns,
                                                   class_names=['Will Not Be Qualified','Will Be Qualified'],
                                                   kernel_width=5)
```

```python
for i in range(900, 2000):
    if(test.loc[[i]].qual_student.item() == 1 and test.loc[[i]].studytime.item() >= 3 and test.loc[[i]].Fedu.item() > 3
      and test.loc[[i]].absences.item() <= 2 and test.loc[[i]].age.item() <= 16):
        print(test.loc[[i]])
```
